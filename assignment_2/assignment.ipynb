### Part 1

```python
def f(x)-> float:
    """
    Returns the float value of the function f(x) = x^3 -x^2 -1 evaluated at x.
    
    Parameters: x- float, the point at which to evaluate f(x)
    Returns: val --- float, the quantity x^3 -x^2 -1
    
    """
    val= x**3 - x**2 - 1
    return val
```

```python
print (f(2), f(3), f(4), f(2.5), f(-1))
```

```python
def df(x)-> float:
    """
    Returns the float value of the derivative of f'(x)= 3x^2 - 2x evaluated at x.
    
    Parameters: x- float, the point at which to evalutate f'(x)
    Returns: val---- float, the quantity 3x^2 - 2x
    
    """
    val = 3*(x**2) - 2*x
    return val
```

```python
print (df(1), df(3), df(4.5), df(-2.3))
```

### Part 2

```python
def newton(f, df, x0, epsilon=1e-6, max_iter=30):
    """
    Returns the number of iterations it took to find the root
    if the value of the function at that point is less than epsilon. 
    Else, returns None upon exceeding the number of iterations allowed
    
    Parameters:
    f----- callable, the function with x as input, its root we are finding
    df----- callable, the derivative of f.
    x0: initial point (float), initial guess of root.
    epsilon: float, the convergence tolerence.
    max_iter: int, the max number of iterations allowed.
    
    Returns: 
    x----- float, the approx root if convergence is there, 
            else None is returned upon max_iter being exceeded.
    """
    
    x= x0 #initialize with the value given as parameter
    for i in range (1, max_iter+1):  #since python index starts at 0 
        x = x - ( f(x) / df(x)) #newton iteration recursive formula
        if abs(f(x)) < epsilon: #condition
            print (f"Found root in {i} iterations")  #print statement
            return x #the value that is almost the root
            
        
    print ("Iteration failed") #else, if we have maximized max_iterations
    return None
```

```python
# testing my function above
f = lambda x: x**2 - 2
df = lambda x: 2*x
newton(f, df, x0=1)
```

```python
# testing my function again
f = lambda x: x**3 - 8
df = lambda x: 3*x**2
newton(f, df, x0=1)
```
